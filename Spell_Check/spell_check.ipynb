```python

from symspellpy import SymSpell,Verbosity
import pkg_resources
import pandas as pd
import dask.delayed as dd 
import dask.dataframe as ddf
from collections import Counter
```

```python

d1 = pd.read_csv("../data/tawiki_data_df.csv")
```

```python
str1 = d1.Content.str.cat(sep=' ')
```

```python
str_counter = Counter(str1.split(' '))
```

```python
str_counter
```

```python
with open("../data/string_dict.txt",'w',encoding='utf-8') as f:
    for key,val in str_counter.items():
        f.writelines(f"{key} {val}\n")
    f.close
```
